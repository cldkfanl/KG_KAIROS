```python
import tensorflow as tf
import pandas as pd

f = 'iris.csv'
df = pd.read_csv(f)
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
print(df.columns)
print(df.shape)
```

```python
#독립변수, 종속변수
inputVar = df[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
tarVar = df[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(inputVar.shape, tarVar.shape)
```

```python
X = tf.keras.layers.Input(shape=4)
Y = tf.keras.layers.Dense(3, activation = 'softmax')(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')
```

```python
model.fit(inputVar, tarVar, epochs = 500)
```

```python
model.predict(inputVar[1:5])
```
