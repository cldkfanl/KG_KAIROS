```python
import tensorflow as tf
import pandas as pd

f = 'boston.csv'
df = pd.read_csv(f)
df.head()
```

```python
print(df.columns)
print(df.shape)
```

```python
#독립변수, 종속변수
inputVar = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
tarVar = df[['medv']]
print(inputVar.shape, tarVar.shape)
```

```python
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(5, activation = 'swish')(X)
H = tf.keras.layers.Dense(3, activation = 'swish')(H)
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'mse')
```

```python
model.fit(inputVar, tarVar, epochs = 5000)
```

```python
model.predict(inputVar[1:5])
```
