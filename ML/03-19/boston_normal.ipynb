```python
import tensorflow as tf
import pandas as pd

f = 'boston.csv'
df = pd.read_csv(f)
df.head()
```

```python
print(df.columns)
print(df.shape)
```

```python
#독립변수, 종속변수
inputVar = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
tarVar = df[['medv']]
print(inputVar.shape, tarVar.shape)
```

```python
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(5)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(3)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(3)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'mse')
```

```python
with tf.device('/GPU:0'):  # 여기서 '0'은 첫 번째 GPU를 의미합니다. 다른 GPU를 선택하려면 해당 번호를 사용하십시오.
    model.fit(inputVar, tarVar, epochs=5000)
```
