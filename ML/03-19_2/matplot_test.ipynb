```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test4.csv')

print(df.columns)
print(df.head(3))
print(df.shape)
```

```python
df_x = np.array(df['x']).reshape(-1,1)
df_y = np.array(df['y']).reshape(-1,1)
df_z = np.array(df['z']).reshape(-1,1)

print(df_x[:3])
print(df_x)
```

```python
np.arange(0,10)
plt.scatter(df_x, df_y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = np.column_stack((df_x, df_y))

```

```python
from sklearn.model_selection import train_test_split
# train_input , test_input , train_output , test_output = train_test_split(data, df_z)

train_input , test_input , train_output , test_output = train_test_split(data, df_z, test_size = 0.2)

train_input , test_input , train_output , test_output = train_test_split(data, df_z, train_size = 0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier , plot_tree
#모델 설정
DT = DecisionTreeClassifier(max_depth = 2)
```

```python
#모델 학습
DT.fit(train_input, train_output)
```

```python
#모델 검증
DT.score(test_input, test_output) # 1.0이면 백점
```

```python
#트리모델 시각화
plot_tree(DT, max_depth = 2, filled = True, feature_names=['X', 'Y'] )
```
