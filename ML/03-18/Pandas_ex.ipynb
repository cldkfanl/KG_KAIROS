```python
#맷플롯립 추가
import matplotlib, matplotlib.pyplot as plt
```

```python
x = df['number']
y = df['height']
```

```python
plt.plot(x) #x축이 number , y축이 number
plt.show()
```

```python
plt.plot(y) #x축이 number , y축이 height
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.axis([0,5,170,180]) #x축은 0~5, y축은 170~180으로 지정(앞 두개는 x 범위, 뒤 두개는 y 범위)
plt.show()
```

```python
plt.plot(x,y, 'r') #색상추가 -> 'k' : black , 'g' : green
plt.show()
```

```python
plt.plot(x,y, 'r*' , x,y+10, 'gs' , x, y+20, 'o' , x, y+30 , 'ko') # 'o', '*', 's' ....는 산점도, 색과 모양을 합쳐서 사용 가능
plt.show()
```

```python
plt.plot(x,y,'r',label = 'height')
plt.plot(x,y+10,'s',label = 'height2')
plt.legend() # 범례 추가
plt.xlabel('number')
plt.ylabel('height')
plt.show()
```

```python
plt.plot(x,y,'g',label='height1')
plt.plot(x,y+10,'rs',label='height2')
plt.legend(loc = 'best', fontsize=20) # 범례의 위치와 크기
plt.xlabel('number')
plt.ylabel('height')
plt.show()
```

```python
plt.scatter(x,y, c='#E389E9',label = 'height1')
plt.scatter(x,y+10, label = 'height2')
plt.legend(loc='best', fontsize=15)
plt.xlabel('number')
plt.ylabel('height')
plt.show()
```
