```python
#섭씨 온도를 사용자로 입력받아 화씨 온도로 변환
def FtoC(temp_f) :
    temp_c = (5.0 * (temp_f - 32.0))/9.0
    return temp_c

temp_f = float(input("화씨온도를 입력해주세요 : "))
print(f"화씨온도 : {FtoC(temp_f)}" )

#소수판별
def is_prime(num) :
    for i in range(2,num) :
        if num % i == 0 :
            return True
    return False

num = int(input(f"검사할 숫자를 입력해주세요 :"))
if is_prime(num) :
    print(f"입력한 수는 소수가 아닙니다.")
else :
    print(f"입력한 수는 소수가 맞습니다.") 


#구 부피 계산
def size(r) :
     V = (4/3) * 3.14 * r**3

     return V

r = int(input("반지름을 입력하세요 :"))
print(f"구의 부피는 {size(r)} 입니다.")
```

```python
from tkinter import *

main = Tk()

main.title("계산기")
main.geometry("500x700")


fir = Entry(main)
fir['font'] = "Arial 20"
fir.pack()

sec = Entry(main)
sec['font'] = "Arial 20"
sec.pack()

output = 0

def plus() :
    num1 = float(fir.get())
    num2 = float(sec.get())
    global output
    output = num1 + num2
    lbl.config(text = output)

def minus() :
    num1 = float(fir.get())
    num2 = float(sec.get())
    global output
    output = num1 - num2
    lbl.config(text = output)

def div() :
    num1 = float(fir.get())
    num2 = float(sec.get())
    global output
    output = num1 / num2
    lbl.config(text = output)

def fow() :
    num1 = float(fir.get())
    num2 = float(sec.get())
    global output
    output = num1 * num2
    lbl.config(text = output)


p_button = Button(main, text="더하기", command=plus)
p_button.pack()

m_button = Button(main, text="빼기", command=minus)
m_button.pack()

b_button = Button(main, text = "곱하기", command=fow)
b_button.pack()

d_button = Button(main, text ="나누기", command=div)
d_button.pack()

lbl = Label(main)
lbl['text'] = output
lbl['font'] = "Arial 20"
lbl.pack()

main.mainloop()
```

```python
def print_menu() :
    print('1. 등록하기')
    print('2. 수정하기')
    print('3. 삭제하기')
    print('4. 목록보기')
    print('5. 종료하기')
    print('----------')

def add_book() :
    name = input("이름을 입력하세요 : ")
    phone = input("전화번호를 입력하세요 : ")
    phoneBook[name] = phone
    print(f"전화번호부에 {name}이 입력되었습니다.")
def mod_book() :
    name = input("수정할 이름을 입력하세요 : ")
    if name in phoneBook :
            phone = input("전화번호를 입력하세요 : ")
            phoneBook[name] = phone
            print(f"전화번호부에 {name}가 수정되었습니다.")
    else :
        print("수정할 이름이 존재하지 않습니다.")


def del_book() :
    name = input("삭제할 이름을 입력하세요 : ")
    if name in phoneBook :
        del(phoneBook[name])
        print(f"전화번호부의 {name}이 삭제되었습니다.")
    else :
        print(f"삭제할 이름이 존재하지 않습니다.")

    
def list_book() :
    if len(phoneBook) == 0 :
        print("출력할 자료가 없습니다.")
    else :
        for i in phoneBook :
            print(f"{i} : {phoneBook[i]}")

    
phoneBook = {}


print_menu()
while True :
    m = int(input("메뉴를 선택하세요 : (숫자만 입력)"))
    if m == 1 :
        add_book()
    elif m == 2 :
        mod_book()
    elif m == 3 :
        del_book()
    elif m == 4 :
        list_book()
    elif m == 5 :
        print("프로그램이 종료되었습니다.")
        break
    else :
        print("잘못된 접근입니다.")


```
