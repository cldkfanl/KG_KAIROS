```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r')       #선그래프 
plt.plot([1,2,3,4], [1, 5, 3, 6], 'o--')       #'b' blue 마커, 'ro' red 원, 'g-' green solid line 등등...

plt.scatter([1, 2, 3, 4], [5, 6, 7, 8])        #산점도
plt.bar([1, 2, 3, 4], [1, 4, 9, 16])        #막대그래프

plt.axis([0,6,0,20])                        #축 설정

plt.figure(figsize= (5,5))                    #배경 크기
plt.show()
```

```python
plt.plot([1,2,3],[4,4,4], '-', color='r', label='Solid')        #다양한 출력 방법 종류
plt.plot([1,2,3],[3,3,3], '--', color='g', label='Dashed')      #실선, 점선 등...
plt.plot([1,2,3],[2,2,2], '-.', color='b', label='Dash-dot')
plt.plot([1,2,3],[1,1,1], ':', color = 'g', label = 'Dotted')

plt.xlabel('x-axis')        #축 이름 설정
plt.ylabel('y-axis')
plt.show()
```

```python
x = [1,2,3]
y = [2,4,5]

plt.plot(x, y, 'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.fill_between(x[1:3] , y[1:3], alpha= 0.5)       #alpha 투명도
plt.show()
```

```python
import numpy as np

x = np.arange(0, 9, 0.5)

plt.plot(x, x, 'ro')
plt.plot(x, x**2, color = '#FFFF00', marker = '*', linewidth = 1)       #색상표를 활용
plt.plot(x, x**3, color = 'forestgreen', marker='^', markersize=5)

plt.axis([0,10,-10,700])

plt.xticks([0,5,10])

plt.yticks(np.arange(1,5))

plt.show()
```

```python
import matplotlib.pyplot as plt

with open ('displ.txt', 'r') as f :
	displ=[]
	for line in f:
		displ.append(float(line))

with open('hwy.txt', 'r') as f:
	hwy=[]
	for line in f:
		hwy.append(float(line))

plt.scatter(displ, hwy)
plt.xlabel("displ")
plt.ylabel("hwy")

plt.show()
```

```python
x = np.arange(3)
years = [2010, 2011, 2012]
domestic = [6801, 6965, 7087]
foreign = [777, 805, 827]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, years)

plt.show()
```
