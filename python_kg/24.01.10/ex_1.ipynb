```python
cap = cv2.VideoCapture(0)           #카메라 사용

while (cap.isOpened()) :
    ret, img = cap.read()           #ret = 프레임 읽기 성공 여부(bool)
    mirror = cv2.flip(img,1)        #mirror 좌우반전
    cv2.imshow('a',mirror)
    cv2.imshow('my win', img)       #img = 프레임 이미지 numpy배열
    if cv2.waitKey(1) & 0xFF == ord('q') :  #q 누르기 전까지
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)           #카메라 사용

cap.set(cv2.CAP_PROP_FRAME_WIDTH,320)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,240)

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
writer = cv2.VideoWriter('test.avi',fourcc, 25.0, (320,240))

while (cap.isOpened()) :
    ret, img = cap.read()           #ret = 프레임 읽기 성공 여부(bool)
    mirror = cv2.flip(img,1)        #mirror 좌우반전
    cv2.imshow('a',mirror)
    writer.write(mirror)
    # cv2.imshow('my win', img)       #img = 프레임 이미지 numpy배열
    if cv2.waitKey(1) & 0xFF == ord('q') :  #q 누르기 전까지
        break
cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
#Line 그리기            param(img = 이미지 파일, start , end , color, thickness)
#cv2.line(param)        circle(img, center, radius, color, thickness)
import numpy as np
import cv2

img = np.zeros((512,512,3), np.uint8)
img = cv2.rectangle(img, (100,100), (400,400), (255,150,150), 7)    #사각형
img = cv2.line(img, (0,0), (511,511), (255,0,0), 10)       #직선
img = cv2.circle(img, (250,250),100, (123,123,123), 5)      #원
pts1 = np.array([[50,50],[200,50],[200,200],[50,200]], np.int32)
pts1 = pts1.reshape((-1,1,2))

pts2 = np.array([[500, 450], [450,450], [450,500], [500,500]], np.int32)
pts2 = pts2.reshape((-1,1,2))
img = cv2.polylines(img, [pts1, pts2], True, (200,200,12), thickness= 10)   #다각형

cv2.putText(img, 'OpenCv', (100,500), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2)    #글씨 삽입
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
drawing = False
mode = True

def draw_circle(event, x, y, flags, param):
    global drawing, mode
    if event == cv2.EVENT_LBUTTONDBLCLK :               #'m' 버튼 클릭시 mode 변경
        if mode :
            cv2.circle(img, (x,y), 30, (0,0,255), -1)   # 원 그리기
        else :
            cv2.rectangle(img, (x-15,y-15), (x+15, y+15), (255,150,150), 7) # 사각형 그리기

    if event == cv2.EVENT_RBUTTONDOWN :                 # 우클릭 상태에 따라 drawing 상태 설정
        drawing = True
    
    if event == cv2.EVENT_RBUTTONUP :
        drawing = False

    if event == cv2.EVENT_MOUSEMOVE :
        if drawing :
            cv2.circle(img, (x,y), 1, (0,0,255), -1)     # drawing 상태에 따라 선 그리기



img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_circle)

while True :
    cv2.imshow('image', img)

    k = cv2.waitKey(1) & 0xFF

    if k == ord('m') :              #m키를 눌러서 모드 변경
        mode = not mode
    elif k == 27 :
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def nothing(x):
    pass

img = np.zeros((300,512,3), np.uint8)
cv2.namedWindow('image')

# trackbar를 생성하여 named window에 등록
cv2.createTrackbar('R', 'image', 0, 255, nothing)
cv2.createTrackbar('G', 'image', 0, 255, nothing)
cv2.createTrackbar('B', 'image', 0, 255, nothing)

switch = '0:OFF\n1:On'
cv2.createTrackbar(switch, 'image', 1, 1, nothing)

while(1):
    cv2.imshow('image', img)

    if cv2.waitKey(1) & 0xFF == 27:
        break

    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G', 'image')
    b = cv2.getTrackbarPos('B', 'image')
    s = cv2.getTrackbarPos(switch, 'image')

    if s == 0:
        img[:] = 0 # 모든 행/열 좌표 값을 0으로 변경. 검은색
    else:
        img[:] = [b,g,r] # 모든 행/열 좌표값을 [b,g,r]로 변경

cv2.destroyAllWindows()
```
